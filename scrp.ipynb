```python
json_convert= json.dumps(result,indent=4)
print(json_convert)
```

```python
from typing import Any
from google_play_scraper import Sort, reviews
from datetime import datetime

class DateTimeEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, datetime):
                return obj.isoformat()
        return super(DateTimeEncoder, self).default(obj)

result, continuation_token = reviews('com.instagram.android',
                                     lang='en',
                                     country='id',
                                     sort=Sort.NEWEST,
                                     count=10, #banyak aplikasi
                                     filter_score_with=None)

result, _ = reviews('com.instagram.android',
                    continuation_token=continuation_token)

json_convert= json.dumps(result,indent=4, cls=DateTimeEncoder) 
```

```python
print(json_convert)
```
